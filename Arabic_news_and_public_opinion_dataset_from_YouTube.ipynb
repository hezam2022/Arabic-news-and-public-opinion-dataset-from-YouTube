**Collect 4500 links**

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd

start=1
end=150

# Set up Chrome options
options = Options()
options.add_argument('--headless')  # Run in headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-gpu')
options.add_argument('--disable-extensions')
options.add_argument('--dns-prefetch-disable')

driver = webdriver.Chrome(options=options)
channel='@France24_ar'
url='https://www.youtube.com/'+str(channel)+'/videos'


unique_links = set()  # Use a set to avoid duplicates
driver.get(url)

# Wait for the page to load
sleep(5)

while True:
    # Scroll down to the bottom
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

    # Wait for new videos to load
    sleep(5)

    script = """
          var links = [];
          var elements = document.querySelectorAll('a#thumbnail');
          elements.forEach(function(element) {
              links.push(element.href);
          });
          return links;
          """

    video_links = driver.execute_script(script)
    # Add new links to the set
    unique_links.update(video_links)
    print('unique links: ',len(unique_links))

    if (start%133)==0:
      data = pd.DataFrame(list(unique_links), columns=['video_link'])
      # Drop rows with null values in the 'video_link' column and reset index
      links = data.dropna(subset=['video_link']).reset_index(drop=True)
      df = pd.DataFrame(columns=['video_link','video_id'])
      df['video_link']=links
      df['video_id'] = df['video_link'].str.extract(r'(?<=v=)([^&]+)')
      print("Original DataFrame shape:", df.shape)

      # Append to CSV
      data.to_excel('/content/drive/MyDrive/news_YouTube/'+str(channel)+'.xlsx', index=False)
      break

    if start==end:
      break

    print("srart: ",start)
    start=start+1
```

**Functions**

```python

def get_video_details(video_id,youtube):

      request = youtube.videos().list(part='snippet,contentDetails,statistics', id=video_id)
      time.sleep(1)
      response = request.execute()

      if response['items']:
          video_info = response['items'][0]
          details = {
              'title': video_info['snippet']['title'],
              'likes': video_info['statistics']['likeCount'],
              'views': video_info['statistics']['viewCount'],
              'date_of_publishing': video_info['snippet']['publishedAt'],
              'hashtags': ', '.join(video_info['snippet'].get('tags', [])),  # Get hashtags if available
              'description': video_info['snippet']['description'],
              'number_of_comments': video_info['statistics'].get('commentCount', 0)  # Get number of comments
          }
          return details
      else:
          return None
from googleapiclient.discovery import build

def get_comments(video_id, youtube):
    comments_data = []
    try:

        request = youtube.commentThreads().list(part='snippet', videoId=video_id, textFormat='plainText', maxResults=100)
        time.sleep(0.2)
        while request:
            response = request.execute()
            for item in response.get('items', []):
                comment = item['snippet']['topLevelComment']['snippet']
                comment_data = {
                    #'author': comment['authorDisplayName'],
                    'comment': comment['textDisplay'],
                    'likes': comment['likeCount'],
                    'replyCount': item['snippet']['totalReplyCount'],  # Number of replies (optional)
                    'publishedAt': comment['publishedAt']
                }
                comments_data.append(comment_data)
            request = youtube.commentThreads().list_next(request, response)  # Get next page of comments
    except HttpError as e:
        print(f"An error occurred: {e}")

    return comments_data
```

```python

from googleapiclient.errors import HttpError
from googleapiclient.discovery import build
from time import sleep
import time

# Replace with your actual API key
#YOUTUBE_V3_API_KEY = '............'
YOUTUBE_V3_API_KEY ='............'
#YOUTUBE_V3_API_KEY = '............'
#YOUTUBE_V3_API_KEY ='............'


youtube = build('youtube', 'v3', developerKey=YOUTUBE_V3_API_KEY)

# Prepare to collect all video details and comments
video_ids = urls['video_id']
all_video_data = []
upperframe=[]


for i, video_id in enumerate(video_ids):

    video_details = get_video_details(video_id,youtube)
    video_comments = get_comments(video_id,youtube)

    # Prepare formatted comments
    comments_details = '\n'.join(
        f"Comment: {comment['comment']} \nLikes: {comment['likes']} \nPublished At: {comment['publishedAt']} \nReply Count: {comment['replyCount']}"
        for comment in video_comments
    )
    # Combine video details and comments into one record
    if video_details:
        final_data = {
            'url': "https://www.youtube.com/watch?v="+str(video_id),
            'id':str(video_id),
            'title': video_details['title'],
            'likes': video_details['likes'],
            'views': video_details['views'],
            'date_of_publishing': video_details['date_of_publishing'],
            'hashtags': video_details['hashtags'],
            'description': video_details['description'],
            'number_of_comments': video_details['number_of_comments'],
            'comments_details': comments_details
        }
        all_video_data.append(final_data)
        print(i)
    if i>3500:
      break
upperframe.extend(all_video_data)
data=pd.DataFrame(upperframe, columns=['url','id','title','likes','views','date_of_publishing','hashtags','description','number_of_comments','comments_details'])
# Save DataFrame to Excel
data.to_csv('/content/drive/MyDrive/news_YouTube/'+str(channel)+'.csv',encoding='utf-8-sig',index=False)
```
